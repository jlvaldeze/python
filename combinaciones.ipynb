```python
#Librerias
import pandas as pd
from itertools import combinations
```

```python
# 1. Leer el archivo CSV
# Reemplaza 'tus_combinaciones.csv' por el nombre real de tu archivo
df = pd.read_csv('Chispazo.csv')

# 2. Filtrar los registros donde la columna 'concurso' sea par
# Se usa el operador % 2 == 0 para verificar si el residuo es cero
df_pares = df[df['CONCURSO'] % 2 == 0].copy()
```

```python
# 3. Mostrar el nuevo DataFrame con los registros pares
print("Registros con concurso par:")
#print(df_pares)
len(df_pares)

# Opcional: Guardar el nuevo DataFrame en un archivo nuevo
# df_pares.to_csv('registros_pares.csv', index=False)
```

```python
df_pares.head(5)
```

```python
# 1. Extraer las combinaciones existentes del DataFrame 'df_pares'
# Supongamos que las columnas de los números se llaman 'n1', 'n2', 'n3', 'n4', 'n5'
# Es importante ordenar cada fila para que la comparación sea efectiva
existentes = set(tuple(sorted(fila)) for fila in df_pares[['R1', 'R2', 'R3', 'R4', 'R5']].values)
```

```python
# 2. Generar el universo de todas las combinaciones posibles (1-28)
todas_posibles = set(combinations(range(1, 29), 5))
```

```python
# 3. Calcular las que NO se tienen (Diferencia de conjuntos)
faltantes = todas_posibles - existentes
len(faltantes)
```

```python
# 4. Resultados
print(f"Total combinaciones posibles: {len(todas_posibles)}")
print(f"Combinaciones encontradas en concursos pares: {len(existentes)}")
print(f"Combinaciones que NO se tienen: {len(faltantes)}")

# Opcional: Convertir a DataFrame para visualizar las primeras 10 faltantes
df_faltantes = pd.DataFrame(list(faltantes), columns=['R1', 'R2', 'R3', 'R4', 'R5'])
print("\nPrimeras 10 combinaciones faltantes:")
print(df_faltantes.head(5))
```

```python
columnas_num = ['R1', 'R2', 'R3', 'R4', 'R5']
primeros_10 = df_pares[columnas_num].head(5).values.flatten()
numeros_en_primeros_10 = set(primeros_10)

print(f"Números presentes en los primeros 10 registros: {numeros_en_primeros_10}")
```

```python
# 2. Identificar números del 1 al 28 que NO están en esos primeros 10 registros
numeros_tot = set(range(1, 29))
numeros_ausentes_iniciales = numeros_tot - numeros_en_primeros_10
numeros_ausentes_iniciales
```

```python
# 3. Filtrar las combinaciones faltantes
# Una combinación se imprime si "alguno de sus números no se tiene en los primeros 10"
# Esto es equivalente a decir: 'si la intersección con numeros_ausentes_iniciales no está vacía'
combinaciones_filtradas = [
    comb for comb in faltantes 
    if any(num in numeros_ausentes_iniciales for num in comb)
]
```

```python
# 4. Mostrar resultados
print(f"Cantidad de combinaciones que cumplen el criterio: {len(combinaciones_filtradas)}")
print("\nEjemplos de combinaciones faltantes con números nuevos:")
for c in combinaciones_filtradas[:15]: # Imprime las primeras 15 para muestra
    print(c)
```

```python
comb_puras_ausentes = list(combinations(numeros_ausentes_iniciales, 5))

# 3. Mostrar resultados
print(f"Total de combinaciones nuevas generadas: {len(comb_puras_ausentes)}")
print("\nPrimeras 10 combinaciones resultantes:")
for c in comb_puras_ausentes:
    print(c)
```

```python

# 1. Calcular la frecuencia de cada número en todo el histórico de df_pares
# 'columnas_num' son ['n1', 'n2', 'n3', 'n4', 'n5']
todos_los_numeros = df_pares[columnas_num].values.flatten()
frecuencias = pd.Series(todos_los_numeros).value_counts(normalize=True).to_dict()

# Asegurar que los números que nunca han salido tengan frecuencia 0 en el diccionario
frecuencia_completa = {i: frecuencias.get(i, 0) for i in range(1, 29)}


```

```python
# 2. Función para calcular el score de una combinación
# Usamos el producto de las probabilidades individuales (Probabilidad Conjunta)
def calcular_score(comb):
    score = 1
    for num in comb:
        score *= frecuencia_completa[num]
    return score

```

```python
# 3. Aplicar el cálculo a 'comb_puras_ausentes'
datos_probabilidad = []
for comb in comb_puras_ausentes:
    prob = calcular_score(comb)
    datos_probabilidad.append({
        'combinacion': comb,
        'score_probabilidad': prob
    })

```

```python
# 4. Crear DataFrame de resultados y ordenar por mayor probabilidad
df_prob = pd.DataFrame(datos_probabilidad)
df_prob = df_prob.sort_values(by='score_probabilidad', ascending=False)

#print("Top 10 combinaciones con mayor peso estadístico histórico:")
#print(df_prob.head(10))
```

```python
# 1. Ordenar de menor a mayor score_probabilidad
df_prob_ordenado = df_prob.sort_values(by='score_probabilidad', ascending=True)

# 2. Configurar pandas para mostrar números decimales con precisión
pd.options.display.float_format = '{:.10f}'.format

# 3. Imprimir el DataFrame
print("Combinaciones ordenadas de MENOR a MAYOR probabilidad histórica:")
print(df_prob_ordenado.head(5))

# Si deseas ver solo las 20 menos probables (las más "frías")
# print(df_prob_ordenado.head(20))
```

```python
# 1. Obtener todas las apariciones de números en df_pares
# Asumiendo que las columnas se llaman R1, R2, R3, R4, R5
columnas_res = ['R1', 'R2', 'R3', 'R4', 'R5']
conteo_numeros = df_pares[columnas_res].values.flatten()

# 2. Calcular la probabilidad individual de cada número (1 al 28)
# Total de números sorteados = total de filas * 5
total_sorteados = len(conteo_numeros)
frecuencias = pd.Series(conteo_numeros).value_counts().to_dict()

# Crear diccionario con P(n) = apariciones / total
# Si un número nunca salió, se le asigna una probabilidad muy pequeña (p.e. 1e-6) o 0
prob_individual = {i: frecuencias.get(i, 0) / total_sorteados for i in range(1, 29)}

# 3. Función para calcular la probabilidad de una combinación
def calcular_prob_combinacion(fila):
    # Producto de las probabilidades de sus 5 números
    return (prob_individual[fila['R1']] * 
            prob_individual[fila['R2']] * 
            prob_individual[fila['R3']] * 
            prob_individual[fila['R4']] * 
            prob_individual[fila['R5']])

# 4. Aplicar al DataFrame de faltantes
# (Asegúrate de que 'df_faltantes' tenga las columnas R1, R2, R3, R4, R5)
df_faltantes['score_probabilidad'] = df_faltantes.apply(calcular_prob_combinacion, axis=1)

# 5. Ordenar de menor a mayor y mostrar
df_faltantes_ordenado = df_faltantes.sort_values(by='score_probabilidad', ascending=True)

print("Combinaciones faltantes con sus probabilidades (ordenadas de menor a mayor):")
pd.options.display.float_format = '{:.12f}'.format
print(df_faltantes_ordenado.head(20))
```

```python
# Definir la combinación buscada
seleccion = [7, 11, 13, 22, 28]

# Filtrar el DataFrame
# Se asegura de que cada columna coincida con el número respectivo
fila_buscada = df_faltantes_ordenado[
    (df_faltantes_ordenado['R1'] == seleccion[0]) &
    (df_faltantes_ordenado['R2'] == seleccion[1]) &
    (df_faltantes_ordenado['R3'] == seleccion[2]) &
    (df_faltantes_ordenado['R4'] == seleccion[3]) &
    (df_faltantes_ordenado['R5'] == seleccion[4])
]

# Verificar si existe y mostrar resultado
if not fila_buscada.empty:
    print("Combinación encontrada:")
    print(fila_buscada)
else:
    print(f"La combinación {seleccion} no se encuentra en el listado de faltantes.")
```

```python
# 1. Definir los números de la combinación
n1, n2, n3, n4, n5 = 7, 11, 13,22, 28

# 2. Filtrar el DataFrame df_pares
# Buscamos la coincidencia exacta en las columnas R1 a R5
busqueda_pares = df_pares[
    (df_pares['R1'] == n1) & 
    (df_pares['R2'] == n2) & 
    (df_pares['R3'] == n3) & 
    (n4 == df_pares['R4']) & 
    (n5 == df_pares['R5'])
]

# 3. Mostrar el resultado
if not busqueda_pares.empty:
    print(f"La combinación {n1, n2, n3, n4, n5} SÍ ha salido en concursos pares:")
    print(busqueda_pares)
else:
    print(f"La combinación {n1, n2, n3, n4, n5} NO se encuentra en el historial de concursos pares.")
```
